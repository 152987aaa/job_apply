```python
import gym
from gridworld import FrozenLakeWrapper

# 创建一个非滑动（is_slippery=False）的FrozenLake环境
env = gym.make("FrozenLake-v0", is_slippery=False)

# 使用FrozenLakeWrapper美化渲染界面
env = FrozenLakeWrapper(env)

# 使用Q-learning算法解决FrozenLake问题
# 初始化Q表，状态和动作的值
Q = {}
for state in range(env.observation_space.n):
    for action in range(env.action_space.n):
        Q[(state, action)] = 0

# 设置学习参数
alpha = 0.5  # 学习率
gamma = 0.9  # 折扣因子
epsilon = 0.1  # 贪婪策略的探索率

# Q-learning算法
num_episodes = 10000
for i in range(num_episodes):
    state = env.reset()
    done = False
    while not done:
        # 贪婪策略选择动作，以epsilon的概率随机探索
        if env.action_space.sample() < epsilon:
            action = env.action_space.sample()
        else:
            action = max(list(action for action in range(env.action_space.n) if Q[(state, action)] == max(Q[(state, action)])))

        # 执行动作，观察下一个状态、奖励和是否结束
        next_state, reward, done, _ = env.step(action)

        # 更新Q值
        old_value = Q[(state, action)]
        next_max = max([Q[(next_state, a)] for a in range(env.action_space.n)])
        Q[(state, action)] = old_value + alpha * (reward + gamma * next_max - old_value)

        # 更新状态
        state = next_state

# 测试策略
success = 0
for _ in range(100):
    state = env.reset()
    while True:
        action = max(list(action for action in range(env.action_space.n) if Q[(state, action)] == max(Q[(state, action)])))
        state, reward, done, _ = env.step(action)
        if done:
            if reward == 1:
                success += 1
            break

# 打印成功率
print("Success rate:", success / 100)
```
