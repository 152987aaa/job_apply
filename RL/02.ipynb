```python
# 只使用官方自带的 CliffWalking-v0 环境，无任何自定义代码
import gymnasium as gym

# 创建环境
env = gym.make("CliffWalking-v1", render_mode="human")

# 重置环境
observation, info = env.reset(seed=0)

print("初始状态:", observation)

# 测试随机走 50 步
for _ in range(50):
    action = env.action_space.sample()  # 随机动作
    obs, reward, terminated, truncated, info = env.step(action)

    print(f"动作: {action} | 状态: {obs} | 奖励: {reward}")

    if terminated or truncated:
        print("到达终点/掉下悬崖，环境重置")
        obs, info = env.reset()

# 关闭环境
env.close()
```

```python
# 只导入最基础的包
import gym

# 创建悬崖环境（自动弹出窗口）
env = gym.make("CliffWalking-v0")

# 重置游戏（必须第一步）
env.reset()

# 循环玩游戏（最简单的交互）
for _ in range(20):
    env.render()        # 显示画面
    action = int(input("输入 0下 1右 2上 3左："))  # 你手动控制
    obs, reward, done, info = env.step(action)    # 走一步

    print("状态：", obs, " 奖励：", reward, " 结束？", done)

    if done:
        print("游戏结束！重置！")
        env.reset()

# 关闭
env.close()
```

```python
import gym
import numpy as np  # 你截图里漏了这个导入

# 创建悬崖行走环境
env = gym.make("CliffWalking-v0")

# 重置环境，获取初始状态
obs = env.reset()

# 循环执行动作直到游戏结束
while True:
    # 随机选择 0-3 之间的一个动作（上下左右）
    action = np.random.randint(0, 4)
    # 执行动作，返回新状态、奖励、是否结束、调试信息
    obs, reward, done, info = env.step(action)
    # 渲染画面（能看见场景）
    env.render()
    # 如果结束（掉下悬崖或到达终点），跳出循环
    if done:
        break

# 关闭环境
env.close()
```
